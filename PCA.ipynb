```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
path = 'C:\\Users\\User\\Desktop\\DATA SCIENCE\\Abhinav R\\datasets\\Attrition_EDA_TOH.csv'
data = pd.read_csv(path, index_col=0)
data.head()
```

```python
len(data.columns)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop('Attrition',axis=1).values
y = data['Attrition']

print(X.shape)
print(y.shape)
```

```python
help(PCA)
```

```python
X = StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components=24)   #specify required no of components
principalComponents = pca.fit_transform(X)
#PCA_X = pd.DataFrame(data = principalComponents)
X= principalComponents
#print(X.head())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 10)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Logistic Regression

logreg = LogisticRegression()

logreg.fit(X_train,y_train)

fold = KFold(n_splits=10, shuffle=True, random_state=10)

accuracies = cross_val_score(estimator=logreg,X=X_train,y=y_train,cv=fold)
print(accuracies)
print("\nKFold")
print(np.mean(accuracies))

y_train_pred = logreg.predict(X_train)
y_train_pred.shape
print("\ntraining")
print(accuracy_score(y_train, y_train_pred, normalize=True, sample_weight=None))

#print(classification_report(y_train, y_train_pred))

#print(confusion_matrix(y_train, y_train_pred))

y_test_pred = logreg.predict(X_test)
y_test_pred.shape
print("\ntest")
print(accuracy_score(y_test, y_test_pred, normalize=True, sample_weight=None))

#print(classification_report(y_test, y_test_pred))

#print(confusion_matrix(y_test, y_test_pred))
```
