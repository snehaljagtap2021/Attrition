```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = 'C:\\Users\\User\\Desktop\\DATA SCIENCE\\Abhinav R\\datasets\\Attrition_EDA_TOH.csv'
data = pd.read_csv(path, index_col=0)

data.head()
```

```python

from sklearn.cluster import KMeans
```

```python
### Clustering with all Parameters


inertia = []
#X = data
for k in np.arange(1,12):
    km = KMeans(n_clusters=k,random_state=10)
    km.fit(data)
    inertia.append(km.inertia_)
```

```python

plt.plot(list(np.arange(1,12)),inertia)
plt.plot(list(np.arange(1,12)),inertia,'rs')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

```python
 
km = KMeans(n_clusters=4,random_state=10)

#fit using fit method
km.fit(data)

#assign labels using predict method on X
labels = km.predict(data)
```

# Hierarchical clustering

```python


from scipy.cluster.hierarchy import linkage
```

```python

link = linkage(data,method='single')
print(link.shape)
```

```python
# link metrix

link
```

```python

from scipy.cluster.hierarchy import dendrogram

#  print dendrogram 
plt.figure(figsize=(20, 10))  
#labelList = range(1, 11)
_ = dendrogram(link,  
           orientation='top')
```

```python
# dendrogram with linkage 'complete'

link_ = linkage(data,method='complete')
print(link_.shape)
plt.figure(figsize=(20, 10))  
#labelList = range(1, 11)
_ = dendrogram(link_,  
           orientation='top')
```

```python

from scipy.cluster.hierarchy import cut_tree


labels_cluster = cut_tree(link_,n_clusters=2)
print(labels_cluster.shape)


set(labels_cluster.reshape(-1,))
```

```python


labels_height = cut_tree(link_,height=8000)
print(labels_height.shape)

set(labels_height.reshape(-1,))
```
